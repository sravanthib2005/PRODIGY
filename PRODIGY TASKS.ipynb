```python
# Import necessary libraries
import os
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder

# Step 1: Generate a synthetic dataset
# We create a dataset with 1000 samples, 10 features, and a binary target
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, n_informative=5, random_state=42)

# Convert the dataset into a DataFrame to mimic real data
feature_columns = [f"feature_{i}" for i in range(X.shape[1])]
data = pd.DataFrame(X, columns=feature_columns)
data['y'] = y  # Add the target column

# Step 2: Explore the synthetic dataset (optional)
print(data.head())
print(data.info())

# Data Preprocessing: Since we generated numeric data, no encoding is needed
# If categorical features existed, we would use LabelEncoder

# Split data into features (X) and target (y)
X = data.drop('y', axis=1)  # Features (all columns except the target 'y')
y = data['y']               # Target (binary classification)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)

# Train the classifier
clf.fit(X_train, y_train)

# Make predictions on the test data
y_pred = clf.predict(X_test)

# Evaluate the model
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
#TASK-04
# Import necessary libraries
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create a sample dataset of social media posts
social_data = pd.DataFrame({
    'message': [
        "This app is fantastic, highly recommended!",
        "The recent update is terrible, it's full of glitches.",
        "It's okay, nothing special but not too bad either.",
        "Customer service was amazing, very helpful!",
        "I hate the new design, it's confusing and buggy.",
        "Good value for the price, not disappointed."
    ]
})

# Step 2: Define a function for sentiment analysis using TextBlob
def analyze_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'

# Apply the sentiment analysis function to the 'message' column
social_data['sentiment_type'] = social_data['message'].apply(analyze_sentiment)

# Step 3: Display the sentiment distribution
print("Sentiment distribution across posts:")
print(social_data['sentiment_type'].value_counts())

# Step 4: Visualize the sentiment distribution using a bar plot
plt.figure(figsize=(8, 6))
sns.countplot(x='sentiment_type', data=social_data, palette='viridis')
plt.title('Distribution of Sentiments in Social Media Posts')
plt.xlabel('Sentiment')
plt.ylabel('Number of Posts')
plt.show()

# Step 5: (Optional) Add a column for polarity score to analyze sentiment intensity
social_data['polarity_score'] = social_data['message'].apply(lambda text: TextBlob(text).sentiment.polarity)

# Step 6: Visualize polarity score distribution using a histogram
plt.figure(figsize=(8, 6))
sns.histplot(social_data['polarity_score'], bins=5, kde=True, color='green')
plt.title('Distribution of Polarity Scores')
plt.xlabel('Polarity Score')
plt.ylabel('Frequency')
plt.show()

# Step 7: Display the processed dataset
print("Processed social media data with sentiment analysis:")
print(social_data)
```

```python
#TASK-05

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap

# Step 1: Generate a sample dataset
np.random.seed(42)

# Create random data for 500 accidents
n_samples = 500

# Simulated time of day (0-24 hours)
time_of_day = np.random.randint(0, 24, size=n_samples)

# Simulated weather conditions
weather_conditions = ['Clear', 'Rainy', 'Snowy', 'Foggy', 'Cloudy']
weather = np.random.choice(weather_conditions, size=n_samples)

# Simulated road conditions
road_conditions = ['Dry', 'Wet', 'Snow', 'Icy', 'Slippery']
road = np.random.choice(road_conditions, size=n_samples)

# Simulated latitude and longitude for a hypothetical city area
latitude = np.random.uniform(low=40.5, high=40.9, size=n_samples)
longitude = np.random.uniform(low=-74.0, high=-73.7, size=n_samples)

# Create the DataFrame
data = pd.DataFrame({
    'time_of_day': time_of_day,
    'weather': weather,
    'road_conditions': road,
    'latitude': latitude,
    'longitude': longitude
})

# Display the first few rows of the dataset
print(data.head())

# Step 2: Data Cleaning (not needed in this case as data is clean)

# Step 3: Analyze and visualize patterns related to time of day

# Convert time_of_day to categorical (Morning, Afternoon, Evening, Night)
data['time_of_day'] = pd.cut(data['time_of_day'], bins=[0, 6, 12, 18, 24], labels=['Night', 'Morning', 'Afternoon', 'Evening'])

# Plot accident frequency by time of day
plt.figure(figsize=(8, 6))
sns.countplot(x='time_of_day', data=data, palette='coolwarm')
plt.title('Accident Frequency by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Accident Count')
plt.show()

# Step 4: Analyze and visualize patterns related to weather conditions

# Plot accident frequency by weather conditions
plt.figure(figsize=(10, 6))
sns.countplot(y='weather', data=data, palette='magma', order=data['weather'].value_counts().index)
plt.title('Accident Frequency by Weather Conditions')
plt.xlabel('Accident Count')
plt.ylabel('Weather Condition')
plt.show()

# Step 5: Analyze and visualize patterns related to road conditions

# Plot accident frequency by road conditions
plt.figure(figsize=(10, 6))
sns.countplot(y='road_conditions', data=data, palette='inferno', order=data['road_conditions'].value_counts().index)
plt.title('Accident Frequency by Road Conditions')
plt.xlabel('Accident Count')
plt.ylabel('Road Conditions')
plt.show()

# Step 6: Visualize Accident Hotspots on a Map

# Create a map centered at the mean latitude and longitude
map_center = [data['latitude'].mean(), data['longitude'].mean()]
accident_map = folium.Map(location=map_center, zoom_start=12)

# Create a list of coordinates for heatmap
accident_locations = data[['latitude', 'longitude']].values.tolist()

# Add HeatMap to the map
HeatMap(accident_locations).add_to(accident_map)

# Save the map to an HTML file (you can open this file to view the map)
accident_map.save("accident_hotspots_map.html")
print("Accident hotspot map saved as 'accident_hotspots_map.html'")

# Step 7: Correlation Analysis for Contributing Factors

# Convert categorical columns to numerical for correlation analysis
data['time_of_day_num'] = data['time_of_day'].cat.codes
data['weather_num'] = pd.Categorical(data['weather']).codes
data['road_conditions_num'] = pd.Categorical(data['road_conditions']).codes

# Select relevant numerical features for correlation
correlation_data = data[['time_of_day_num', 'weather_num', 'road_conditions_num']]

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Factors')
plt.show()
```
